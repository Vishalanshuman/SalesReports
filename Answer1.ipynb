```python
df['Product_Category_2']=df['Product_Category_2'].fillna(0)
df['Product_Category_3']=df['Product_Category_3'].fillna(0)
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["Purchase"],bins=20,kde=True)
plt.title("Purchase Items Distribution")
plt.xlabel('Purchase Items')
plt.ylabel('Frequency')
plt.show()
```

```python
# Getting Outliers
plt.figure(figsize=(10,6))
sns.boxplot(df['Purchase'])
plt.xlabel("Purchase Itmes")
plt.title("BoxPlot of Purchase")
plt.show()
```

```python
# Removing Outliers
df=df[df['Purchase']<20000]

plt.figure(figsize=(10,6))
sns.boxplot(df['Purchase'])
plt.title('Box Plot of Purchase After Removing Outliers')
plt.xlabel("Purchase Items")
plt.show()
```

```python
# Analysis By Gender
gender_analysis=df.groupby('Gender')['Purchase'].mean()
gender_analysis
```

```python
# Analysis By Marital Status
marital_status_analysis=df.groupby('Marital_Status')['Purchase'].mean()
marital_status_analysis
```

```python
#analysis By Occupation
occupation_analysis=df.groupby('Occupation')['Purchase'].mean()
occupation_analysis
```

```python
# Purchase By City Category Analysis
purchase_by_city=df.groupby('City_Category')['Purchase'].sum()
purchase_by_city
```

```python
# Purchase by Age Group
age_group_analysis=df.groupby('Age')['Purchase'].sum()
age_group_analysis
```

```python
# Drop Unnecessary colums
df.drop(['Product_ID'],axis=1)
```

```python
# Converting Categorical Data into integer using map function
df['Gender']=df['Gender'].map({"M":1,"F":0})
df.head()
```

```python
df.rename(columns={
    "User_ID":"UserID",
    "Product_ID":"ProductID",
    "Gender":"IsMale",
    "Age":"AgeGroup",
    "Occupation":"OccupationID",
    "City_Category":"CityCategory",
    "Stay_In_Current_City_Years":"YearsInCity",
    "Marital_Status":"IsMarried",
    "Product_Category_1":"Category1",
    "Product_Category_2":"Category2",
    "Product_Category_3":"Category3",
    "Purchase":"PurchaseTotal"
},inplace=True)
df.head()
```

```python
# Age vs Purchase Barplot
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='PurchaseTotal',y='AgeGroup')
plt.title('Age vs Purchase Total')
plt.show()
```

```python
# Occupation Vs Purchased Items

plt.figure(figsize=(10,6))
sns.barplot(data=df,x='OccupationID',y='PurchaseTotal')
plt.title('Occupatoin Vs Purchased Producsts Barplot')
plt.ylabel('Purchase Total')
plt.xlabel('Occupation Id')
plt.show()
```

```python
# Purchase items vs Category 1
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x="Category1",y="PurchaseTotal",hue='IsMale')
plt.title("Purchase Total Vs Category1")
plt.xlabel("Category 1")
plt.ylabel("Purchase Total")
plt.show()
```

```python
# Purchase Items vs Category 2
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='Category2',y='PurchaseTotal',hue='IsMarried')
plt.title('Category 2 and Purchase Total Plot')
plt.ylabel("Purchase Items")
plt.show()
```

```python
# Purchase Items vs Category 3
plt.figure(figsize=(10,6))
sns.jointplot(x='PurchaseTotal',y='Category3',data=df,kind='reg',
              truncate=False,xlim=(0,60),ylim=(0,12),color="m",height=7)
plt.xlabel('Purchased Total')
plt.show()
```

```python
# City Category Pie Chart

plt.figure(figsize=(10,6))
df['CityCategory'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['lightblue','lightgreen','lightcoral'])
plt.title('City Category Distribution')
plt.ylabel('')
plt.xlabel('City Category')
plt.show()
```
